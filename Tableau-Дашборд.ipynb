# Проект: История TED-конференций.

# Описание проекта

**TED (от англ. technology, education, design — «технологии, образование, дизайн») — некоммерческий фонд, который проводит популярные конференции. На них выступают специалисты из разных областей и читают лекции на актуальные социальные, культурные и научные темы.**

# Описание данных

tableau_project_data_1.csv

tableau_project_data_2.csv

tableau_project_data_3.csv

tableau_project_event_dict.csv

tableau_project_speakers_dict.csv

**Файлы tableau_project_data_1.csv, tableau_project_data_2.csv, tableau_project_data_3.csv хранят данные выступлений. У них одинаковая структура:**

talk_id — идентификатор выступления;

url — ссылка на запись выступления;

title — название выступления;

description — краткое описание;

film_date — дата записи выступления;

duration — длительность в секундах;

views — количество просмотров;

main_tag — основная категория, к которой относится выступление;

speaker_id — уникальный идентификатор автора выступления;

laughter_count — количество раз, когда аудитория смеялась в ходе выступления;

applause_count — количество раз, когда аудитория аплодировала в ходе выступления;

language — язык, на котором велось выступление;

event_id — уникальный идентификатор конференции.

**Файл tableau_project_event_dict.csv — справочник конференций. Описание таблицы:**

conf_id — уникальный идентификатор конференции;

event — название конференции;

country — страна проведения конференции.

**Файл tableau_project_speakers_dict.csv — справочник авторов выступления. Описание таблицы:**

author_id — уникальный идентификатор автора выступления;

speaker_name — имя автора;

speaker_occupation — профессиональная область автора;

speaker_description — описание профессиональной деятельности автора.

# Выводы:

**История выступлений:**
1) Больше всего выступлений проходило в США, Канаде и Великобритании.

2) Наиболее популярны следующие категории: технологии, общество, наука, глобальные проблемы и культура.

3) В 2020,2021 году состав категорий включал в себя технологии, общество и глобальный проблемы.

**Тематики выступлений:**
1) Наиболее популярны следующие категории: Наука, Технологии, Культура.

2) В целом по популярности тематики в США и Канаде очень схожи. Есть незначительные отличия, например в США на втором месте находятся две тематики - технологии и другое. В то врем как в Канаде технологии занимают только третью позицию.

3) Чаще всего вызывают аплодисменты категория развлечения и социальные изменения. Реже всего дизайн и наука. Больше всего аплодировали выступлению Sebastian Wernicke в тематике развлечения.

4) Чаще всего вызывают смех категории образовние и бизнес. Реже всего - социальные изменения и глобальные проблемы. Самым смешным оказалось выступление Einstein the Parrot в тематике образования.

5) Явной зависимости между длительностью видео и количеством просмотров нет. Убивают ли школы креативность? - видео с наибольшим количеством просмотров(72 млн.). Самое длительное видео - Три секрета успеха компании Нетфликс(62 мин.).

**Авторы выступлений:**
1) Топ три основных области деятельности авторов это: Писатель, журналист, художник.

2) У 90% авторов всего одно выступление. Чаще всего выступал Hans Rosling.

3) Из дизайнеров чаще всего выступал Tom Wujec. Меньше всего смотрели его выступление Learn to use the 13th-century astrolabe.

**Доп.дашборд ДРУГИЕ:**
1) По полученной таблице заметно, что с 2019 по 2021 годы количество тематик в категории другое уменьшается. А в 2021 вовсе остается всего три категории: Здоровье, Экономика и Города.

2) Из первого выводы следует, что уменьшается и количество самих выступлений. Если в 2019 максимально было 7-8 выступлений, то уже в 2021 всего по одному выступлению каждой из категорий.
